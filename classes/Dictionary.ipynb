# Dictionary :
- Dictionary is ordered collection key-value (i.e Items) pairs. It is mutable and defines with '{}'.
## Dictionay Format :
- {key : value}.
## Characteristics :
- Dictionary is ordered collection.
- Dictionary is mutable.
- Dictionary does not support indexing.
- Dictionary keys are unique.
- Dictionary values allows duplicates.
- List can't be used as dictionary key, because unhashable.
- Dictionary values can access by using keys.

```python
# Ordered Collection
details = {"Name" : "Raju","Course" : "Python","Batch" : 57,"Age" : 21}
print(details)
```

```python
# Dictionary values access by using keys
# Updating dicttionary values
# Adding new items to dictionary
details = {"Name" : "Raju","Course" : "Python","Batch" : 57,"Age" : 21}
print(details['Course'])                     # if the key does not exists it will raise an key error.
print(details["Age"])
details['Age'] = details["Age"]+1            #Updating
print(details["Age"])
details['Location'] = "Hyderabad"            #Adding the key
print(details)      
```

```python
details = {"Age" : 25,"Name" : "Raju","Course" : "Python","Batch" : 57,"Age" : 21}             # It will take the last value
print(details)
details = {"Name" : "Raju","Course" : "Python","Batch" : 57,"Marks":[20,30,40],"Age" : 21}
print(details)
```

## Dictionary Methods :
 get(key, default_value:optional()) : It returns the value if key is exists in dictionary, otherwise it returns default value

```python
# Dictionary keys are unique
details = {"Name" : "Raju","Course" : "Python","Batch" : 57,"Age" : 21}
print(details["Name"])
print(details["Marks"])                                        # It will raise key error
print(details.get('Marks'))                                    # It will return the none because we didnt mentioned default_value
print(details.get('Marks','Marks not available'))              # It will not raise an error if there is no specified column
```

- Update(new_dictionary) : This method updates the values new_dictionary keys already exists in the dictionary, otherwise it adds new items to dictionary

```python
details = {"Name" : "Raju","Course" : "Python","Batch" : 57,"Age" : 21}
details.update({'Batch' : 58,'Course' : 'Java','Location' : "Hyd"})
print(details)
```

- `Keys()` : This method returns the list dictionary keys
- `values()` : This method returns the list dictionary values
- `Items()` : This method return the list of tuple of key-value pairs.

```python
details = {"Name" : "Raju","Course" : "Python","Batch" : 57,"Age" : 21}
print(details.keys())
print(details.values())
print(details.items())
```

## Item Removing Methods : 
- popitem() : It return and removes the last item as tuple
- pop(key) : This method return the  specified key item and return the specified keys value, otherwise `Key Error`

```python
details = {"Name" : "Raju","Course" : "Python","Batch" : 57,"Age" : 21}
item = details.popitem()
print(item)
print(details)
print(details.pop('Name'))
print(details)
```

```python
# Membership - It checks only keys not values
details = {"Name" : "Raju","Course" : "Python","Batch" : 57,"Age" : 21}
print('Raju' in details)
print('Name' in details)
```
