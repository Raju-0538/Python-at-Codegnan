```python
from datetime import datetime
current_timestamp = datetime.now()     # to get the current timestamp
print(current_timestamp)
today = current_timestamp.date()       # to extract the date from currentstamp
print(today)
print(datetime.now().time())           # to get the current time or we can use current_timestamp also
```

```python
# Extracting the day,month,year from date
print(current_timestamp.date().day)
print(datetime.now().date().month)
print(current_timestamp.date().year)
```

```python
# Extracting the hour,minute,second from time
print(current_timestamp.time().hour)
print(datetime.now().time().minute)
print(current_timestamp.time().second)
```

```python
# change date format
print(today.strftime("%d / %m / %y"))
print(today.strftime("%d / %m / %Y"))
print(today.strftime("%d / %b / %Y"))
print(today.strftime("%d / %B / %Y"))
print(today.strftime("%A / %b / %Y"))
print(today.strftime("%a / %b / %Y"))
print(today.strftime("%d-%a / %b / %Y"))
```

```python
# change time format
curr = datetime.now().time()
print(curr.strftime("%H : %M : %S"))
print(curr.strftime("%I : %M  %p"))
```

```python
# implementing timer
import time
from IPython.display import clear_output
h_count = m_count = s_count = 0
while True:
    print(f"{h_count} : {m_count} : {s_count}")
    time.sleep(1)
    s_count += 1
    if s_count == 60:
        m_count += 1
        s_count = 0
        if m_count == 60:
            h_count += 1
            m_count = 0
            if h_count == 24:
                h_count = 0
    clear_output(wait=True)
```
