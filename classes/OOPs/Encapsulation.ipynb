# Encapsulation : 

            Encapsulation means, wrapping the methods and attributes in single unit that one is class. the main aim of encapsulation is to restrict the direct access of attributes and methods.

            - This type of restriction can implement by using access modifiers.

            - types of Access Modifiers :

                        1. Public
                        2. Protected
                        3. Private

            - In python we dont have any keywords to represent it as publis, protected and private but by using underscore( _ ) can define.

### 1. Public :

            The public members(attributes, methods) does not starts with underscore and this members can access anywhere in the program.

### 2. Protected :

            the protected member starts with single underscore. This members can access in inside the class and in sub classes also. but we cant access outside the class.

### 3. Private :

            the private member starts with double underscore( __ ). This members can only access in inside the class. we cant access in subclasses or outside the class.

### Name Mangling :

            we can use name mangling for access protect and private members outside the class.

            syntax :

                    protected - obj._attribute
                    private - obj._classname__attribute

```python
class Bank:
    def __init__(self,name,account,password,balance):
        self.name = name
        self.account = account
        self.__password = password
        self._balance = balance
    def __total_users(self):    # getter meethod
        return "Total users : 101"
    def get_balance(self):    # getter method
        return f"Current Balance : {self._balance}"
    def deposit(self,deposit_amount):    #setter method
        self._balance += deposit_amount
        return f"{deposit_amount} Deposited successfully !"

u1 = Bank("Ravi",1001,"Ravi@123",2000)
print(u1.name)
print(u1.get_balance())
print(u1.deposit(1000))
print(u1.get_balance())
print(u1._balance)      # Accessing the protected member outside the class
print(u1._Bank__password)    # Accessing the private member outside the class
```
