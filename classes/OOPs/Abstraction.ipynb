# Abstraction :

        It Hides the comlex implementation and the functionality, Those are implemented in child class.

```python
from abc import ABC,abstractmethod
class Mobile(ABC):
    @abstractmethod
    def ram(self):
        pass
    @abstractmethod
    def rom(self):
        pass
    def display(self, Hzs):
        return f"This Mobile display quality is {Hzs}"

class Oppo(Mobile):
    def ram(self):
        return "this is from Oppo class and The Ram is 4 GB"
    def rom(self):
        return "this is from Oppo class and The ROM is 128 GB"

class Redmi(Mobile):
    def ram(self):
        return "this is from Redmi class and The Ram is 12 GB"
    def rom(self):
        return "this is from Redmi class and The ROM is 264 GB"

# mo = Mobile()              # Can't instantiate abstract class Mobile without an implementation for abstract methods 'ram', 'rom'
oppo = Oppo()
redmi = Redmi()
print(oppo.display(100))
print(redmi.display(200))
print(oppo.ram())
print(redmi.ram())
print(oppo.rom())
print(redmi.rom())
```
