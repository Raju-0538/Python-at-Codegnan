# Methods : 
      The function definition is implemented inside the class those functions called as methods. 

      types of Methods :

                  1. Instance Method
                  2. class method
                  3. static method

### 1. Instance Method : 

        The method definition takes first parameter as 'self' those methods called as Instance Method. This method only works on Instance Attributes.

        Syntax : 
                    class class_name :
                        def method_name(self,parameters):
                            # block code
                            

### 2. class Method : 

        The method definition startss with @classmethod and takes first parameter as 'cls' those methods called as class Method. This method only works on class Attributes.

                Syntax : 
                    class class_name :
                        @classmethod
                        def method_name(cls,parameters):
                            # block code


### 3. static Method : 

        The method definition starts with @staticmethod and it does not take first parameter as self or cls those methods called as static methods. This methods works on except class and instance attributes.

                Syntax : 
                    class class_name :
                        @staticmethod
                        def method_name(parameters):
                            # block code

### Initialization Constructor :

        This constructor is used to initialize the instance attributes data. It takes first parameter as 'self' and it executes whenever the object is created.

            Syntax : 
                    class class_name :
                        
                        def __init__(self,parameters):
                            # Initialization


```python
# class attribute - school
# instamce attribute - id, name,class

class Students:
    school = 'XYZ School'   # Class Attribute
    def __init__(self,Id,name,Class):
        self.Id = Id   # Instance Attribute
        self.name = name   # Instance Attribute
        self.Class = Class   # Instance Attribute
    def details(self):  # Instance method
        return f"My name is {self.name} and my id is {self.Id} and my class is {self.Class} from {Students.school}"
    # class Method
    @classmethod
    def update(cls, new_school):
        cls.school = new_school
        return "Updated successfully"
    @staticmethod
    def percentage(marks:list):
        return (sum(marks)/300)*100
name = input("Enter name : ")
Id = int(input("Enter Id : "))
Class= input("Enter Class : ")
obj = Students(Id,name,Class)
print(obj.details())    
```
