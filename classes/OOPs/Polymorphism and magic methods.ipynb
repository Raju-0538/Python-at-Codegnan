# Polymorphism :

            In python 'polymorphism' poly means many and morphism means forms(behaviour).

            - In polymorphism single function it shows different behaviour based on input context.

            - In python polymorphism can achieve through the following ways :

                        1. through operators
                        2. polymorphism through functions
                        3. polymorphism through methods:
                                  |--> 1. Runtime polymorphism (method overriding)
                                  |--> 2. compiletime polymorphism (method overloading)

### 1. Polymorphism through Operators :

            Here single operator shows different actions based on the input context.

            Ex:
                    + -> input(int) -> addition operation (3+5 = 8)
                         input(string) -> String concatination ('Ra'+'ju' = 'Raju')

                    * -> input(int) -> Multiplication operation (5*5 = 25)
                         input(string) -> String Repetition ('Hi'*3 = 'HiHiHi')

### 2. polymorphism through functions :

            Here a function shows different behaviour based on input context.

            Ex:

                    len() -> input(str) -> returns string length (len('Hello') = 5)
                             input(List) -> returns 

### 2. polymorphism through Methods :

            Here a Method shows different behaviour based on input context.
            Again it is devided into 2 types :

                        1. Method Overriding(Runtime Polymorphism)
                        2. Method Overloading

####       1. Method Overriding(Runtime Polymorphism):

                        A method implemented in child class that is already existed in parent class with the same name and it is possible through inheritance and methods names should be same. Whenever if you called that method with child class object the child class method overrides the parent class method it is known as method overriding .

                        Ex:
                            class A:
                                def show()

                            class B(A):
                                def show()

                            obj = B()
                            obj.show()         

####       2. Method Overloading(compilation Polymorphism):

                        A single class contains multiple methods with the same name in that case the last defined method will overloads all previous methods is known as Method Overloading. Method Overloading is not possible directly in python, but we can achieve same functionality through the default parameters or variable length positional or keyword arguments.

                        Ex:
                            class A:
                                def show()

                                def show()

                                def show()

                            obj = A()
                            obj.show()         

```python
# Method Overriding
# Parent class method without any parameter
class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender
    def display(self):
        return f"My name is {self.name} and iam from Person class"

class Student(Person):
    def __init__(self, name, age, gender, stdid, Class):
        super().__init__(name,age,gender)
        self.stdid = stdid
        self.Class = Class
    def display(self):
        return f"My name is {self.name} and iam from Student class and my Roll number is {self.stdid}"

b = Student("Raju",22,"Male",538,"B.Tech")
b.display()
```

```python
# Method Overriding
# Parent class method with parameter
class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender
    def display(self,location):
        return f"My name is {self.name} and iam from Person class and iam from {location}"

class Student(Person):
    def __init__(self, name, age, gender, stdid, Class):
        super().__init__(name,age,gender)
        self.stdid = stdid
        self.Class = Class
    def display(self):
        return f"My name is {self.name} and iam from Student class and my Roll number is {self.stdid}"

b = Student("Raju",22,"Male",538,"B.Tech")
b.display()
b.display("Hyderabad")
```

```python
# Method Overloading
# without default arguments
class Math:
    def add(self,a,b):
        return a+b
    def add(self,a,b,c):
        return a+b+c
    def add(self,a,b,c,d):        # This method overloads the previous 2 methods.
        return a+b+c+d

b = Math()
# b.add(1,2)              # It raises error of missing 2 arguments.
# b.add(1,2,3)            # It raises the error like missing 1 argument
b.add(1,2,3,4)
```

```python
# Method Overloading
# with default arguments
class Math:
    # def add(self,a,b):
    #     return a+b
    # def add(self,a,b,c):
    #     return a+b+c
    def add(self,a,b,c = 0,d = 0):        # This method overloads the previous 2 methods.
        return a+b+c+d

b = Math()
print(b.add(1,2))             # It will not raises error of missing 2 arguments.
print(b.add(1,2,3))            # It will not raises the error like missing 1 argument
print(b.add(1,2,3,4))
```
