# Inheritance : 

            In python, A class get properties from another class this process called as 'Inheritance'.

            - A child class is inherited from parent class.

            Types of Inheritance :

                        1. single inheritance
                        2. multi level inheritance
                        3. multiple inheriatance
                        4. Hierarichal inheritance
                        5. Hybrid inheritance

###         1. Single Inheritance :

                        A child class get properties from single parent class. this is called as single inheritance.

                        Syntax :

                                class parent_class_name :
                                            # attributes
                                            # methods
                                class child_class (parent _class_name):
                                            # attributes
                                            # methods

###         2. multi level Inheritance :

                        A class(child class) get properties from some another class (parent class), that class also get properties from some another class (grand parent class). this is called as multi level inheritance.

                        Syntax :

                                class grnad_parent_class_name :
                                            # attributes
                                            # methods
                                class parent_class_name (Grnad_parent_classs) :
                                            # attributes
                                            # methods
                                class child_class (parent _class_name):
                                            # attributes
                                            # methods

###         3. multiple Inheritance :

                        A single class(child class) get properties from multiple classes (parent classes). this is called as multiple inheritance.

                        Syntax :

                                class parent_class_name1 :
                                            # attributes
                                            # methods
                                class parent_class_name2 :
                                            # attributes
                                            # methods
                                class child_class (parent_class_name1,parent_class_name2):
                                            # attributes
                                            # methods

###         4. Hierarichal Inheritance :

                        multiple classes(child classes) get properties from single class (parent class). this is called as Hierarichal inheritance.

                        Syntax :

                                class parent_class_name :
                                            # attributes
                                            # methods
                                class child_class_name1 (parent_class_name):
                                            # attributes
                                            # methods
                                class child_class_name2 (parent_class_name):
                                            # attributes
                                            # methods

###         5. Hybrid Inheritance :

                        It is a combination of two or more types of inheritances. this is called as Hybrid inheritance.

```python
# 2. multi level Inheritance

class Student:
    def __init__(self, name, age, stdid,gender):
        self.name = name
        self.age = age
        self.stdid = stdid
        self.gender = gender
    def details(self):
        return f"My name is {self.name}, My student id is {self.stdid} and iam {self.age} years old."
class College(Student):
    def __init__(self, name, stdid, marks,age,gender):
        super().__init__(name,age,stdid,gender)
        self.marks = marks
    def displaymarks(self):
        return f"My name is {self.name}, My student id is {self.stdid} and I got {self.marks} Marks."
class Engineering(College):
    def __init__(self, name,age,gender,stdid,marks,college,cgpa):
        super().__init__(name, stdid,marks,age,gender)
        self.college = college
        self.cgpa = cgpa
    def displaycollege(self):
        return f"My name is {self.name} My college name is {self.college} and my cgpa is {self.cgpa}"
s = Engineering('Raju',21,'Male',538,555,"SASI",8.1)
print(s.details())
print(s.displaymarks())
print(s.displaycollege())
```

```python
# 3. Multiple Inheritance
class Personal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def details(self):
        return f"My name is {self.name} and my age is {self.age}"

class Education:
    def __init__(self, stdid, college):
        self.stdid = stdid
        self.college = college
    def details1(self):
        return f"My student ID is {self.stdid} and iam from {self.college} college"
class Information(Personal, Education):
    def __init__(self, name, age, stdid, college, marks):
        super().__init__(name, age)
        Education.__init__(self,stdid, college)
        self.marks = marks
    def info(self):
        return f"My name is {self.name}, Iam {self.age} Years old, Mystudent ID is {self.stdid}, My college name is {self.college} and I got {self.marks} marks."

s = Information(name = "Raju",age = 22,stdid = 538,college = "SASI",marks = 555)
print(s.info())
print(s.details())
print(s.details1())
```

```python
# 4. Hierarichal Inheritance
class Mobile : 
    def __init__(self, shape, purpose):
        self.shape = shape
        self.purpose = purpose
    def det(self):
        return f"The shape of the mobile is {self.shape} and it is used for {self.purpose}."
class Mi(Mobile):
    def __init__(self, shape, purpose, name, model):
        super().__init__(shape,purpose)
        self.name = name
        self.model = model
    def det1(self):
        return f"The mobile is from Brand {self.name} and model is {self.model}. The shape of the mobile is {self.shape} and it is used for {self.purpose}."
class Oppo(Mobile):
    def __init__(self, shape, purpose, name, model,price):
        super().__init__(shape,purpose)
        self.name = name
        self.model = model
        self.price = price
    def det2(self):
        return f"The mobile is from Brand {self.name} and model is {self.model}. The shape of the mobile is {self.shape} and it is used for {self.purpose} and the price is {self.price}."

s1 = Mi("Square","Browsing","Redmi","Note 13 5G")
s2 = Oppo("Square","Browsing","Oppo","Reno 6 5G",2000)

print(s1.det())
print(s1.det1())
print(s2.det2())
```
