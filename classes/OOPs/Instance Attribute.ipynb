## 1. Class :

        A class is a blueprint of a real world object or template of object.

        - Class is a collectioin of attributes(data) and methods(i.e behaviour).

        Syntax : 

                class class_name:
                    # attributes
                    # Methods

## 2.Objects :

        Object is a instance of a class and it is a real world item.

                # object creation
                    obj = class_name(arguments)

        1. Attribute: 

                The variables are defined inside the class those variables called as attributes. The attributes are used for storing the data.

        Types of Attribute :

                1. Class attributes
                2. Instance attributes

            1.Instance Attributes :

                    The variables are inside the initialization constructor those variables called as instance attributes.

                    - The instance attributes data is different to every object.

            2. Class Attributes :

                    The variables are defined outside of initialization constructor those variables called as class attributes.

                    - The class attributes data is same to all object.

                

        

```python
# define person class 
class Student :
    school = 'XYZ school'
    def __init__(self,name,age):   # initialization constructor( Magic method)
        self.name = name
        self.age = age
    def details(self):
        return f"My name is {self.name} and iam a {self.age} years old"

s1= Student("Ravi",20)
s2 = Student("Janu",19)
print(s1.name)
print(s2.age)
print(s1.details())
```

```python
# define person class 
# updating instance attributes data
class Student :
    school = 'XYZ school'
    def __init__(self,name,age):   # initialization constructor( Magic method)
        self.name = name
        self.age = age
    def details(self):
        return f"My name is {self.name} and iam a {self.age} years old"

s1= Student("Ravi",20)
s2 = Student("Janu",19)
print(s1.name)
s1.name = "Raju"
print(s1.name)
print(s2.age)
print(s1.details())
s1.school = "ABC School"
print(s2.school,s1.school)
```
