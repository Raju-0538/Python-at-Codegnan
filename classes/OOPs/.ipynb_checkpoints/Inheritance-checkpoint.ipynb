
            Pyhton does not support high security like java. here, we can access the protected member and private member in outside the class by using following ways :
            1. To access the protected member obj._protectedmembername
            2. To access the private member by using name mangling - obj._classname.__privatemembername

            - The advantage of encapsulation is it can provide the security for data as well as for functionalities.

# Inheritance : 

            In python, A class get properties from another class this process called as 'Inheritance'.

            - A child class is inherited from parent class.

            Types of Inheritance :

                        1. single inheritance
                        2. multi level inheritance
                        3. multiple inheriatance
                        4. Hierarichal inheritance
                        5. Hybrid inheritance

###         1. Single Inheritance :

                        A child class get properties from single parent class. this is called as single inheritance.

                        Syntax :

                                class parent_class_name :
                                            # attributes
                                            # methods
                                class child_class (parent _class_name):
                                            # attributes
                                            # methods

###         2. multi level Inheritance :

                        A class(child class) get properties from some another class (parent class), that class also get properties from some another class (grand parent class). this is called as multi level inheritance.

                        Syntax :

                                class grnad_parent_class_name :
                                            # attributes
                                            # methods
                                class parent_class_name (Grnad_parent_classs) :
                                            # attributes
                                            # methods
                                class child_class (parent _class_name):
                                            # attributes
                                            # methods

###         3. multiple Inheritance :

                        A single class(child class) get properties from multiple classes (parent classes). this is called as multiple inheritance.

                        Syntax :

                                class parent_class_name1 :
                                            # attributes
                                            # methods
                                class parent_class_name2 :
                                            # attributes
                                            # methods
                                class child_class (parent_class_name1,parent_class_name2):
                                            # attributes
                                            # methods

###         4. Hierarichal Inheritance :

                        multiple classes(child classes) get properties from single class (parent class). this is called as Hierarichal inheritance.

                        Syntax :

                                class parent_class_name :
                                            # attributes
                                            # methods
                                class child_class_name1 (parent_class_name):
                                            # attributes
                                            # methods
                                class child_class_name2 (parent_class_name):
                                            # attributes
                                            # methods

###         5. Hybrid Inheritance :

                        It is a combination of two or more types of inheritances. this is called as Hybrid inheritance.


```python
# single Inheritance
# Without child class implementation
# when child class dont have constructor, in that case parent class constructor behaves like child class constructor
class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender
    def details(self):
        return f"My name is {self.name}. Iam a {self.age} Years old."
class Student(Person):
    pass

s = Student('Raju',21,'Male')
print(s.details())
```

```python
# single Inheritance
# With child class implementation
# when child class have constructor it will behave like child class constructor, parent class constructor behaves like parent class constructor. so we can pass the values to child class constructor
class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender
    def details(self):
        return f"My name is {self.name}. Iam a {self.age} Years old and gender is {self.gender}."
class Student(Person):
    def __init__(self,stdid,name,age,branch):
        self.name = name
        self.stdid = stdid
        self.age = age
        # self.gender = gender
        self.branch = branch

s = Student(538,'Raju',21,'CSE')
print(s.name)
# print(s.gender)    # It rises attribute error because student reference does not contains gender attribute
print(s.age)
```

```python
# single Inheritance
# With child class implementation
# when child class have constructor it will behave like child class constructor, parent class constructor behaves like parent class constructor. so we can pass the values to child class constructor
class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender
    def details(self):
        return f"My name is {self.name}. Iam a {self.age} Years old and gender is {self.gender}."
class Student(Person):
    def __init__(self,stdid,name,age,gender,branch):
        # self.name = name
        super().__init__(name,age,gender)
        self.stdid = stdid
        # self.age = age
        # self.gender = gender
        self.branch = branch

s = Student(538,'Raju',21,'Male','CSE')
print(s.name)
print(s.gender)    
print(s.age)
print(s.details())
```
