# Conditional Statements :
                The purpose of conditional statement is to make decisions.
    
    Types :
                
                1. if 
                2. if - else
                3. if -elif -else
                4. Nested if
                
    1. If statement :

                The block code is executes when if condition evaluates to TRUE.
                
                Syntax :
                if condition:
                                #block of code.

    2. If-else :

                The if block code is executes when condition evaluates to TRUE. The else block code executes when condition evaluates to FALSE.

                Syntax :
                if condition :
                                #block of code.
                else :
                                #block of code.
                

```python
# check whether the number is divisible by 3 or not
num = int(input())
if((num%3) == 0) :
    print(f"{num} is divisible by 3")
else:
    print(f"{num} is not divisible by 3")
```

```python
# check whether the number is divisible by both 3 and 5
num = int(input())
if((num%3) == 0 and (num%5) == 0) :
    print(f"{num} is divisible by 3 and 5")
else:
    print(f"{num} is not divisible by 3 and 5")
```

### 3. if -elif -else :

                The if block code executes when if condition evaluates to true, otherwise it sequentially checks with elif conditions , if any conditon is true, them it executes respective block otherwise else block will executes.

                note : If any block executed immediately the execution goes to main block.       

```python
# check whether the number is positive, negitive or zero
num = int(input())
if num == 0:
    print(f"{num} is zero")
elif num < 0:
    print(f"{num} is Negative")
else:
    print(f"{num} is Positive")
```

```python
# divisible by 3 - Fizz
# divisible by 5 - Buzz
# both - FizzBuzz
num = int(input())
if num%3==0 and num%5==0:
    print("FizzBuzz")
elif num%3 == 0:
    print("Fizz")
elif num%5 == 0:
    print("Buzz")
else:
    print("Number is not divisible 3 or 5")
```

```python
# check whether the number is positive even, negative even, positive odd , negative odd or zero
# Using elif method
num = int(input())
if(num%2==0 and num>=0):
    print("Given number is positive even")
elif(num%2==0 and num<0):
    print("Given number is negative even")
elif(num%2==1 and num>0):
    print("Given number is positive Odd")
else:
    print("Given number is negative odd")
```

```python
# check whether the number is positive even, negative even, positive odd , negative odd or zero
# Using elif method loop optimization
num = int(input())
if(num%2==0 and num>=0):
    print("Given number is positive even")
elif(num%2==0):
    print("Given number is negative even")
elif(num>0):
    print("Given number is positive Odd")
else:
    print("Given number is negative odd")
```

```python
# check whether the number is positive even, negative even, positive odd , negative odd or zero
# using nested loops
num = int(input())
if num%2 == 0:
    if num>=0 :
        print("Given number is positive even")
    else:
        print("Given number is negative even")
else:
    if num>=0 :
        print("Given number is positive odd")
    else:
        print("Given number is negative odd")
```

```python
# find the biggest number among the 3 numbers
a = int(input())
b = int(input())
c = int(input())
if a>b and a>c:
    print(a)
elif b>c:
    print(b)
else:
    print(c)
```
