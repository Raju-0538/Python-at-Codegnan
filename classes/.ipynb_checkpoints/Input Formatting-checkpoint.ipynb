# Input formatting

```python
# reading integer from user
num = int(input("Enter a number"))
print(type(num))
print(num)
```

```python
# reading float from user
num = float(input("Enter a number"))
print(type(num))
print(num)
```

```python
# reading complex from user
num = complex(int(input()),int(input()))
print(type(num))
print(num)
```

```python
# reading boolean from user
num = bool(input())
print(type(num))
print(num)
num = bool(int(input()))
print(type(num))
print(num)
```

```python
# reading string from user
string = input()
print(type(string))
print(string)
```

```python
# reading list of integers from user
# map() : This function applies the functionality on each element in iterable
# Syntax : map(fun,iterable)
lst = list(map(int,input().split( )))      #If we dont use the list it will return the object location
print(type(lst))
print(lst)
mul = lst[2]*5
print(mul)
```

```python
# reading list of strings from user
# map() : This function applies the functionality on each element in iterable
# Syntax : map(fun,iterable)
lst = list(input().split())      #If we dont use the list it will return the object location
print(type(lst))
print(lst)
con = lst[1]+lst[2]
print(con)
```

```python
# reading tuple from user
tup = tuple(map(int,input().split()))
print(type(tup))
print(tup)
print(tup[3]*5)
```

```python
# Output formatting
name = 'Raju'
age = 20
print("My name is",name,"and iam",age,"Years old")                       # Using comma
print("My name is "+name+" and iam "+str(age)+" Years old")
print("My name is %s and iam %d Years old"%(name,age))                   # using access specifiers
print("My name is {} and iam {} Years old".format(name,age))             # Using .format (dot format)
print(f"My name is {name} and iam {age} Years old")                      # Using f Format
# print(f"My name is {} and iam {} Years old")                           # Raises an error
```
