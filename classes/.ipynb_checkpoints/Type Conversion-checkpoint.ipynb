# Type Conversion :
                Converting one datatype format into another datatype format. this is called as type conversion
        These are 2 types :
        
                    1. implicit type conversion
                    2. explicit type conversion(type casting)
                    
    1. Impicit typeconversion :

                    - The python interpreter automatically converts the datatype because it prevents the data loss
                    - Implicit type conversion can happen in arithmetic operators
                    - Ex : 3.5+5, 3/1,....

    2. Explicit Typeconversion(Typecasting) :
    
                    - Here typeconversion happens by using the python builtin datatypes.
                    - Ex : int(3.5),...

## Python Builtin Datatype Functions :

                    1. integer - int()
                    2. float - float()
                    3. complex val - complex()
                    4. boolean - bool()
                    5. string - str()
                    6. list - list()
                    7. tuple - tuple()
                    8. set - set()
                    9. dictionary - dict()
            

```python
# int to float,bool,string
num = 10
print(float(num))
print(bool(num))
print(str(num))
print(complex(num))
```

```python
# float to int, boolean and string
val = 10.2556
num = int(val)
print(type(num),num)
print(bool(val))
print(type(val))
string = str(val)
print(val,type(string))
print(string[:4], string[-1])
```

```python
# Sting to int,float,bool,list,tuple,set
# if string value is int
s = '18238718'
print(int(s))
print(float(s))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
```

```python
# Sting to int,float,bool,list,tuple,set
# if string value is float
s = '18238718.2465'
print(int(float(s)))
print(float(s))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
```

```python
# Sting to bool,list,tuple,set
# if string value is characters
s = 'Raju'
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
```

```python
# List to tuple, string, set
lst = [1,2,3,'hi',2,3,4]
tup = tuple(lst)
string = str(lst)
set_val = set(lst)
print(tup)
print(string, type(string))
print(set_val)
```

```python
# tuple to list, string, set
tup = (1,2,3,'hi',3,4,5,2)
print(list(tup))
print(str(tup))
print(set(tup))
```

```python
# dictionary to string,list, set, tuple
dic = {'Name' : 'Raju','Age' : 21,'Location' : 'Hyderabad'}
print(str(dic),type(str(dic)))
print(list(dic))
print(set(dic))
print(tuple(dic))
```

```python
# set to string,list,tuple
a = {1,2,32,3,2,2,5,'hi','hello'}
print(type(str(a)),str(a))
print(type(list(a)),list(a))
print(type(tuple(a)),tuple(a))
```

```python
# list to dictionary conversion
lst = [('a',1),('b',2)]
print(lst)
print(type(lst))
```
