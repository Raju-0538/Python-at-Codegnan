```python
def print_n(x):
    if x == 0:
        return 1
    print_n(x-1)
    print(x,end =" ")
a = int(input())
print_n(a)
```

```python
n = int(input())
res = ""
while n>0:
    rem = n%2
    res = str(rem)+res
    n = n//2
print(res)
```

```python
a = input()
res = ""
for i in a:
    res = i+res
print(res)
```
