# LISTS :
 - List is a ordered collection different data type elements. Lists are mutable and defines with '[]'.
## Characterstics:
 - Lists are ordered collection.
 - Lists are mutable.
 - List allows indexing and slicing.
 - List contains duplicate elements.
 - List is Hetrogeneous datatype.
 - List is unhashable datatype.
 - EX : [1,3,5,4,5,'Hello',2,1,3]

```python
lst = [1,2,3.3,"hi",1,3,5]
print(lst)
```

```python
print(lst[4])
```

```python
#slicing
lst1 = lst[1:4]
print(lst1)
print(lst[-3:11])
print(lst[::-1]) #reverse the list.
```

```python
l1 = []
l2 = list()
print(type(l1),type(l2))
```

# List operations:
- List concatination (using + opeartor)
- List Repetition (usong * operator)

```python
l1 = [1,2,3]
l2 = [3,4,5,6]
print(l1+l2)
print(l1*3)
print([0]*10)
print(l2[2]*6)
```

# List Methods :
## Item adding methods:
  - append(item) : It adds item at end of list
  - extend(iterable) : It adds the all iterator elements at end of list
  - insert(pos,item) : It adds item at specified position

```python
l = [1]
l.append(100)
l.append([200,300])
l.extend((1,2,3))
l.insert(0,1000)
print(l)
```

## Item removing methods : 
- remove(item) : It removes the first occurance item from the list if it is exists in the list otherwise 'Value Error'.
- pop(pos) : It remove and returnn the specified position item if position is passed, otherwise it remove and returns the last item fromlist
- clear() : It removes all elements from list.

```python
lst = [10,20,30,20,40,50]
lst.remove(30)
val2 = [(lst.pop(3))]
print(lst)
print(val2)
val2.extend((50,50,60,70))
print(val2)
val2.clear()
print(val2)
```

## Index and count()
- index(item): If item is present in list it returns first occurance of the index otherwise "value error"
- count(item) : It returns the count of total occurances of item

```python
lst = [10,20,30,20,40,50]
ind = lst.index(20)
print(ind)
print(lst.index(50))
print(lst.count(20))
```

## Sorting items
- sort(reverse = false) : By default it sorts the original list into Ascending order. if reverser is true it sorts in Descending order.

```python
lst = [1,4,2,5,-5,3,-45,-4,30]
lst.sort()
print(lst)
lst.sort(reverse = True)
print(lst)
```

```python
lst = ['hi','hello','banti']
lst.sort()
print(lst)
lst.sort(reverse = True)
print(lst)
```

```python
#Builtin Functions - len(),min(),max(),sum()
lst = [1,2,3,4,5]
print(len(lst),min(lst),max(lst),sum(lst))
avg = sum(lst)/len(lst)
print(avg)
```

```python
lst = ['hi','hello','abi']
print(len(lst))
print(len(lst),min(lst),max(lst))
```
