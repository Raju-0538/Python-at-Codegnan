```python
# 1 to n
a = int(input())
for i in range(1,a+1):
    print(i)
```

```python
# Even Numbers
a = int(input())
count = 0
for i in range(1,a+1):
    if i%2 == 0:
        print(i)
        count += 1
if count < 1:
        print("-1")
```

```python
# Even, Odd, Positive and Negative
a = int(input())
b = list(map(int, input().split()))
even = 0
odd = 0
pos = 0
neg = 0
for i in range(0,len(b)):
    if int(b[i])%2 == 0:
        even += 1
    else:
        odd += 1
    if int(b[i])>0:
        pos += 1
    elif b[i]<0:
        neg += 1
print(f"Even: {even}")
print(f"Odd: {odd}")
print(f"Positive: {pos}")
print(f"Negative: {neg}")
```

```python
#  Fixed Password
while True:
    x = int(input())

    if x == 1999:
        print("Correct")
        break
    else:
        print("Wrong")
```

```python
# Max
n = int(input())
arr = []
for i in range(n):
    arr.append(int(input()))
print(max(arr))
```

```python
# Multiplication table
a = int(input())
for i in range(1,13):
    print(f"{a} * {i} = {a*i}")
```

```python
# Factorial
a = int(input())
arr = []
for i in range(a):
    arr.append(int(input()))
for j in arr:
    count = 1
    for k in range(j,1,-1):
        count *= k
    print(count)
```

```python
# One Prime
a = int(input())
count = 0
for i in range(2,a):
    if a%i == 0:
        count += 1
if count <1 and a>=2:
    print("Yes")
else:
    print("No")
```

```python
# Palindrome
a = input()
rev = ""
rev = a[::-1]
if a == rev:
    print(rev)
    print("YES")
else:
    print(int(rev))
    print("NO")
```

```python
# Primes from 1 to n
a = int(input())
for i in range(2,a+1):
    count = 0
    for j in range(2,i):
        if i%j == 0:
            count += 1
    if count >= 1:
        continue
    else:
        print(i)
```

```python
# Divisors
a = int(input())
for i in range(1,a+1):
    if a%i == 0:
        print(i)
```

```python
# GCD
a, b = map(int, input().split())

while b != 0:
    a, b = b, a % b

print("GCD =", a)
```
