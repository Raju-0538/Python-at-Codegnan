# Regular Expressions :

        

```python
import re
```

```python
# match
string = "The sky is blue and ocean is blue"
res = re.match("The",string)
print(res)
print(res.span())     # It returns the matched pattern position
print(res.group())    # It returns the matched result
```

```python
# match
try:
    string = "The sky is blue and ocean is blue"
    res = re.match("sky",string)
    print(res)
    print(res.span())     # It raises an error
    print(res.group())    # It raises error
except:
    print("There is no matched pattern")
```

```python
# search
try:
    string = "The sky is blue and ocean is blue"
    res = re.search("blue",string)
    print(res)
    print(res.span())
    print(res.group())
except:
    print("There is no match found")
```

```python
# search
try:
    string = "The sky is blue and ocean is blue"
    res = re.search("Mlue",string)
    print(res)
    print(res.span())
    print(res.group())
except:
    print("There is no match found")
```

```python
# findalll
try:
    string = "The sky is blue and ocean is blue"
    res = re.findall("blue",string)
    print(res)
    # print(res.span())
    # print(res.group())
except:
    print("There is no match found")
```

```python
# findalll
try:
    string = "The sky is blue and ocean is blue"
    res = re.findall("[abc]",string)
    print(res)
    # print(res.span())
    # print(res.group())
except:
    print("There is no match found")
```

```python
# split
string = "The sky is blue and ocean is blue"
res = re.split("blue",string)
print(res)
```

```python
# split
string = "The sky is blue and ocean is blue"
res = re.split("[blue]",string)
print(res)
```

```python
# split
string = "The sky is blue and ocean is blue"
res = re.split("[zx]",string)
print(res)
```

```python
# validate mobile number
try :
    mobile = input("Enter mobile number : ")
    result = re.fullmatch("^[6-9][0-9]{9}",mobile)
    # print(result)
    if result:
        print(f"{mobile} is valid number !")
    else:
        print(f"{mobile} is not a valid number !")
except:
    print("There is no matched pattern")
```

```python
# validate pan number
try :
    PAN = input("Enter PAN number : ")
    result = re.fullmatch("^[A-Z]{5}[0-9]{4}[A-Z]$",PAN)
    # print(result)
    if result:
        print(f"{PAN} is valid number !")
    else:
        print(f"{PAN} is not a valid number !")
except Exception as e:
    print(f"There is no matched pattern {e}")
```

```python
# validate Aadhar number
try :
    mobile = input("Enter Aadhar number : ")
    result = re.fullmatch(r"^[2-9][0-9]{3}\s?[0-9]{4}\s?[0-9]{4}",mobile)
    # print(result)
    if result:
        print(f"{mobile} is valid number !")
    else:
        print(f"{mobile} is not a valid number !")
except:
    print("There is no matched pattern")
```

```python
# validate company email
try :
    mobile = input("Enter your Email : ")
    result = re.fullmatch(r"^[a-z0-9][0-9a-z.]+\@[0-9a-z]+\.[a-z]{2,3}$",mobile)
    # print(result)
    if result:
        print(f"{mobile} is valid email !")
    else:
        print(f"{mobile} is not a valid email !")
except:
    print("There is no matched pattern")
```

```python
# validate Date
try :
    mobile = input("Enter your Date : ")
    result = re.fullmatch(r"^(0?[1-9]|[12][0-9]|3[01])/(0?[1-9]|1[0-2])/(\d{4})$",mobile)
    # print(result)
    if result:
        print(f"{mobile} is valid Date !")
    else:
        print(f"{mobile} is not a valid Date !")
except:
    print("There is no matched pattern")
```
