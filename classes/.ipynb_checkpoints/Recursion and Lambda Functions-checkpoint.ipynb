# Recursion :

        A function calls itself either directly or indirectly.

        Types of Recursions :

                    1. Direct recursion
                    2. Indirect recursion

##  1. Direct Recursion :

            A function calls itself until it reaches to base condition. 

    - Base condition : 

                The base condition is used to stop the infinite recursive call.

    - Recursive call :

                The function calls recursively until it reaches to base condition.

```python
def fun(x):
    if x == 0:
        return 0
    fun(x-1)
    print(x)
fun(3)
```

```python
def fun(x):
    if x == 0:
        return 0
    print(x)
    fun(x-1)
fun(3)
```

```python
# factorial of a number
def fact(x):
    if x==0 or x==1:
        return 1
    return x*fact(x-1)
fact(5)
```

```python
# find the sum of digits
def sum(x):
    if x == 0:
        return 0
    rem = x%10
    return rem + sum(x//10)
sum(1234)
```

```python
# find the sum of digits
def sum(x):
    if x == 0:
        return 0
    rem = x%10
    return 1 + sum(x//10)
sum(123415267674)
```

# Lmabda functions (or) Anonymous functions :

                - The function does not have function name those funtions called as ananymous functions. 
                - The ananymous functions can implemented by using 'lambda' keyword.
                
                Syntax :

                        lambda arguments : Expression

## Single line if else :

        True_exp if condition else False_exp

```python
# find the square of a number using lambda function
s = lambda x : x**2
print(s(10))
```

```python
# find the square of a number using lambda function
arr = list(map(int, input().split()))
print(list(map(lambda x : x**2, arr)))
```

```python
# find the even numbers in a list using lambda function
arr = list(map(int, input().split()))
print(list(filter(lambda x : x%2 == 0, arr)))
```

```python
x = 10
res = 'even' if x%2 == 0 else 'odd'
print(res)
```

```python
# checking whether the number is even or odd using lambda function
arr = list(map(int, input().split()))
print(list(map(lambda x : f"{x}-even" if x%2 == 0 else f"{x}-odd",arr)))
```
