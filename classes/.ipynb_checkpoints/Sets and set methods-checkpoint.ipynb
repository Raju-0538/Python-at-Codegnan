# Sets :
            Sets is unordered collection of different datatype unique elements. Sets are mutable and defines with '{}'.
## Characteristics :
- set is unordered collection.
- set elements are unique.
- set does not supports indexing.
- sets are mutable.
- set allows hashable datatypes as elements.
- set elements can be used as dictionary keys.
- List cant be used as set elements.

```python
# Unordered collection
s = {1,2,3,2,3,'hi','a',2.02,(10,5,20,30)}
print(s)
```

```python
# List cant be used as set element
s = {1,2,3,2,3,'hi','a',[10,20,'hi'],2.02,(10,5,20,30)}
print(s)
```

## Set Methods

```python
# Element adding method
# add(ele) : It adds the element to set if element npt in set, otherwise nothing
s = {1,2,3,4,5,4,3}
print(s)
s.add(100)
s.add(5)
s.add("Code")
print(s)
```

### Element removing method
- remove(item) : It removes the item from set if it exists in set, otherwise it raises `Key Error`
- discard(item) : It removes the item from set if it exists in set, otherwise it raises Nothing

```python
s = {1,2,3,4,5,4,3}
s.remove(5)
# s.remove(100)           #Key error
print(s)
s.discard(3)
s.discard(100)           #Does not raises any error
print(s)
```

# Sets Operations
- Union(set) : It returns the all unique elements from all sets

```python
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {1,2,8,9}
print(s1.union(s2,s3))
print(s1 | s2 | s3)
```

```python
# Intersection(set) : It returns all common elements
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {1,2,8,9}
print(s1.intersection(s2,s3))
print(s1 & s2 & s3)
print(s1.intersection(s2))
print(s1 & s3)
```

```python
# Intersection_update(set) : It returns all common elements and updates the set
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {5,1,2,8,9}
print(s1.intersection_update(s2,s3))
print(s1)
```

```python
# difference : It returns the set those are not in another set
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {5,1,2,8,9}
print(s1.difference(s2,s3))
print(s1 - s2 - s3)
print(s1.difference(s2))
print(s1 - s3)
print((s1 | s2) - s3)
print(s1 | s2 - s3)
```

```python
# difference_difference : It returns the set those are not in another set and updates the set
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {5,1,2,8,9}
s1.difference_update(s2,s3)
print(s1)
```

```python
# Symmetric difference : It returns all the elements in the list except common elements
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {5,1,2,8,9}
# print(s1.symmetric_difference(s2,s3))
print(s1^s2^s3)
print(s1.symmetric_difference(s2))
print(s1^s3)
```

```python
# Symmetric difference : It returns all the elements in the list except common elements
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {5,1,2,8,9}
# print(s1.symmetric_difference(s2,s3))
s1.symmetric_difference_update(s2)
print(s1)
```

```python
# issubset(set)
# issuperset(set)
#isdisjoint(set)
a = {1,2,34,4,5,6,7}
b = {34,4,5}
c = {10,20,30}
print(a.issubset(b))
print(b.issubset(a))
print(a.issuperset(b))
print(b.issuperset(a))
print(a.isdisjoint(b))
print(a.isdisjoint(c))
print(b.isdisjoint(c))
print(dir(a))
```
