```python
n = int(input())
count = 0
while(n>0):
    if n%2 == 0:
        n /= 2
        count += 1
    else:
        n -= 1
        count += 1
print(count)
```

```python
n = int(input("Enter no of rows and columns: "))
mat = []

for i in range(n):
    row = list(map(int,input().split()))
    mat.append(row)
    print()

count = 0
for i in range(n):
    count += mat[i][i]+mat[i][n-i-1]
print(count)
```
