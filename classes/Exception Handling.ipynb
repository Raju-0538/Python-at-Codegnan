## Exception : 

        Exception is nothing but, it is an error whenever an exception is occurs it stops the program execution and it returns type of exception and followed by exception message.

        Types of Exceptions :

                1. compile time errors or Syntax errors
                2. Runtime errors
                3. Logical Errors.

###    1. Compile time or syntax errors:

                The syntax errors raise during the compilation time of the program. Those errors we can call it as compile time errors or syntax errors. We cant handle the syntax errors.

                Ex:

                    Indentation errors,syntax errors.

###    2. Runtime errors:

                The Runtime errors raise during the Runtime time of the program. Those errors we can call it as Runtime time errors. We can handle the runtime errors by using Exception handling blocks (try and except blocks).

                Ex:

                    Index Errors, Type errors, key error,etc.

####   Exception Handling Blocks:

                Syntax : 
                        try:
                            # block of code that may contains exception
                        except:
                            # exception handling block
                        else (Optional):
                            # This block executes when there is no exception raised in try block
                        finally (Optional):
                            # This block always executes even exception raise or not

```python
# Try block with exception
try:
    lst = [1,2,3]
    print(lst[10])
    print("hi")
except:
    print("Something wrong in try block")
else:
    print("There is no error in try block")
finally:
    print("Program done")
```

```python
# Try block without exception
try:
    lst = [1,2,3]
    print(lst[1])
    print("hi")
except:
    print("Something wrong in try block")
else:
    print("There is no error in try block")
finally:
    print("Program done")
```
