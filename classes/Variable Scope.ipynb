## Scope :

        A scope represents region of a program. and where variable or name can access.

### Types of scopes :

            1. local scope
            2. enclosing scope
            3. global scope
            4. builtin scope

### 1. Local scope :

            defining a variable inside the function.

### 2. Enclosing scope:

            A variable is defined in outer function and accessing in inner function(Nested functions).

### 3. Global scope :

            A varible is defined outside the all functions and using in functions.

### 4. Builtin scope :

            The python provides the builtin functions like print(), input(), len(), etc.

    - Note:

            whenever if you try to access any variable or name it starts checks in 'LEGB' Rule.

                        L - Local scop
                        E - Enclosing scope
                        G - Global scope
                        B - Builtin scope

```python
# 1. local scope
def local():
    x = 10
    print(x)
local()
```

```python
# 2. enclosing scope
def outer():
    x = 10              #enclosing variable
    def inner():
        y = 20
        print(x,y)     # Local variable
    inner()
outer()
```

```python
# 3. Global scope
x = 10                      # Global scope variable
def outer():
    y=20                    # Enclosing scope variable 
    def inner():
        z = 30              # Local scope variable
        print(x,y,z)        # Print() is builtin scope
    inner()
outer()
```

```python
x = 10
def outer():
    y = x+1
    print("Local scope variable",y)
outer()
print("Global scope variable",x)
```

```python
# updating the global variable in local scope by using global keyword
# If we update the global varible in local scope using global keyword it will update in global scope also
x = 10
def outer():
    global x
    x += 1
    print("Local scope variable",x)
outer()
print("Global scope variable",x)
```

```python
# updating enclosing variable in local scope directly will raise an error
def outer():
    x = 10              #enclosing variable
    def inner():
        x += 20
        print(x)     
    inner()
outer()
print(x)
```

```python
# updating enclosing variable in local scope
# by using non local keyword
x = 40
def outer():
    x = 10              #enclosing variable
    def inner():
        nonlocal x
        x += 20
        print(x)     
    inner()
outer()
print(x)
```
