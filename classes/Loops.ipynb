# Loops :

        Performs same operation repeatedly again and again this called as loop.

    Types of Loops :

                    1. for loop
                    2. while loop

## 1. For loop :

            for loop will use when you know total iterations and this loop can use iterate over the list,tuple,set,dictionary ,string and range() function.
            - This is automatically increment loop
            
            Synatx : for iterator in iteable:
                        #block code
                     else:(Optional):
                         #block code

                    iteable = list/set/tuple/dictionary/String/range()

## 2. While loop :

            - The loop body executes until condition evaluate to false.
            - In while loop manual increment/decrement.
            - If u dont know the total number of iterations, we will use while loop.

            Syntax : # Initialization
                     while condition:
                            # block code
                            # Increment/decrement
                     else : (Optional)
                             # block code

```python
# with loops
lst = list(map(int, input().split()))
for i in lst:
    if i%2 == 0:
        print(i**2)
    else:
        print(i+1)
```

## Loop control statements : 

                    1. continue
                    2. break
                    
### 1. Continue :

            After executing of 'continue' keyword execution immediately goes to next iteration and it skips the current iteration after continue keyword statements.

### 2. break :

            After executing of 'break' keyword immediately execution goes to out of loop.
            - that means loop will terminates.
            - The loops else block executes when break keyword not executed.

```python
for i in range(1,11):
    if i%2 == 0:
        continue
    print(i)
```

```python
a = list(map(int, input().split()))
num = int(input())
n = len(a)
for i in range(0,n):
    if num in a:
        if a[i] == num:
            print(i)
            break
    else:
        print("-1")
        break
```

# Nested Loops :
