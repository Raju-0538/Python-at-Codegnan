```python
# Comparison
a = input()
eq = a.find("=")
if eq == -1:
    if eval(a):
        print("Right")
    else:
        print("Wrong")
else:
    if int(a[:eq-1])==int(a[eq+2:]):
        print("Right")
    else:
        print("Wrong")
    
```

```python
# Mathematical Expression
a = input()
eq = a.find("=")
x = int(eval(a[:eq-1]))
if x == int(a[eq+2:]):
    print("Yes")
else:
    print(eval(a[:eq-1]))
```
