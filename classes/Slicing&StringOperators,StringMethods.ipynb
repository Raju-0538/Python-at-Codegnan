## String Methods
- Case conversion methods
  - Upper() : It converts entire string into upper case
  - Lower() : It converts entire string into lower case
  - Title() : It cinverts the first letter of every word into upper and remining lower
  - Swapcase() : It converts uppercase into lowercase and lower into upper
  - Capitalize() : It converts first character into uppper case and remining into lower case

```python
s = "the Sky is Blue and Ocean is blue"
print(s.upper())
print(s.lower())
print(s.title())
print(s.swapcase())
print(s.capitalize())
```

- Remove white space or pattern at end of string
   - strip() : By default it removes the white space at both ends of string
   - lstrip() : It removes the white space at left side of string
   - rstrip() : It removes the white space at right side of string


```python
s = "         codegnan            "
print(s)
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
s = "@@@@codegnan @@@@@"
print(s)
print(s.strip('@'))
print(s.lstrip('@'))
print(s.rstrip('@'))
```

```python
# split() : It splits the entire string into list of substrings based on the specified seperator.
# buy default it splits at space
s = "the Sky is Blue and Ocean is blue"
print(s.split())
print(s.split("is"))
print(s.split("e"))
```

```python
# ' sep'.join(iterable) : It joins all the iterable elements with the separator
w = ['the', 'Sky', 'is', 'Blue', 'and', 'Ocean', 'is', 'blue']
print("".join(w))
print(" ".join(w))
print("-".join(w))
```

### substring index finding
- index(substring) : If substring is exists in string, it returns the first occurence of substring position otherwise 'Value Error'
- find(substring) : If substring is exists in string, it returns the first occurence of substring position otherwise '-1'

```python
s = "the Sky is Blue and Ocean is blue"
print(s.index('is'))
print(s.index('z'))
```

```python
s = "the Sky is Blue and Ocean is blue"
print(s.find('is'))
print(s.find('z'))
print(s.find('is',10)) #It starts checking from 10th position
print(s.find('s',10))
print(s.find('e',-5))
```

```python
# count(substring) : It returns the total non-overlapping occurances of substring
s = "ABCBCBCbbaabb"
print(s.count("B"))
print(s.count("BCB"))
print(s.upper().count("B"))
print(s.upper().count("z"))
```

```python
# replace(old_substrin, new_substring) : It replace the old substring with the new substring in string
s = "the Sky is Blue and Ocean is blue"
print(s.replace("Blue","Cloudy"))
print(s.replace("e","E"))
print(s.replace("e","E",2))   #It replaces first 2 occurances only.
print(s.replace("Ble","Cloudy"))
```

### String Checking Methods
- String checking methods returns the boolean value as result
- isupper() : It returns True, If the entire string is in upper case
- islower() : It returns True, If the entire string is in lower case
- isalpha() : It returns True, If the entire string contains alphabets
- isdigit() : It returns True, If the entire string contains digits
- isalnum() : It returns True, If the entire string contains either alphabets or digits
- statrtswith(substring) : It returns True, If the string starts with substring
- endswith(substring) : It returns True, If the string ends with substring

```python
s = "the Sky is Blue and Ocean is blue"
s2 = "chaeckingwithputspaces"
print(s.isupper())
print(s.islower())
print(s.isalpha())
print(s2.isalpha())
print(s.isdigit())
print(s.isalnum())
print(s2.isalnum())
print('1623'.isalnum())
print(s.startswith("the"))
print(s.startswith("blue"))
print(s.endswith("blue"))
print(s.endswith("and"))
print(s.startswith("green"))
```
