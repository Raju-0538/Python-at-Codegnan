```python
# immuatable(length and position of remaining elements is remains same.)
t = (1,2,3,[10,20,30],4,5)
t[3].append(100)
print(t)                 #After operation it doesn't change the length of object and doesn't change the remaining elements position
```

```python
# empty tuple representation
# single element representation
t1 = ()
tuple = (3,)
print(tuple)
lst = [10]
print(type(t1),type(tuple),type(lst))
```

```python
# Tuple operations
# Tuple concatination
# Tuple repetition
t = (1,2,34.5,'hello',3)
t1 = (10,20,30)
print(t+t1)
print(t1*4)
```

```python
# Tuple Methods : 
# index()
# count()
t = (1,2,34.5,'hello',3)
t1 = (10,20,30,10,20,40,10)
print(t.index(3))
print(t1.count(10))
print(t1.count(20))
```

```python
# Tuple Builtin functions
t1 = (10,20,30,10,20,40,10)
print(len(t1),min(t1),max(t1),sum(t1))
```
