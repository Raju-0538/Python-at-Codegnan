# Modules :

    Types of Modules :

            1. User defined Modules
            2. Builtin Modules

### 1. Userdefined Modules :

            The modules are defined by the user or programmer , those modules are called as Userdefined modules.

### 2. Builtin Modules :

            The modules which are provided by the python software is called as Builtin modules.

            Ex:

                - Math                          -> Advanced Mathematical operations
                - random                        -> To generate random numbers
                - os                            -> File or Directory creation
                - sys                           -> Related to Interpreter
                - pandas                        -> Data Manipulation purpose
                - numpy                         -> Numerical calculations
                - matplot lib & seaborn         -> Data Visualization
                - scikit learn                  -> For Machine Learning Models

```python
import math,random,os,sys
```

```python
print(math.ceil(4.25))    # It returns the upperbound integer value
print(math.floor(4.25))   # It returns the lowerbound integer value
print(math.factorial(5))  # It returns the factorial of a number
print(math.gcd(10,6,4))   # It returns the GCD of a number
print(math.lcm(10,6,4))   # It returns the LCM of a number
print(math.pow(10,2))     # It returns the float type of {a} power {b} value
print(math.sqrt(9))       # It will return the square root value of a number
print(math.log10(2))      # It will return  the log10 value of a number
```

```python
print(random.random())     # It generates the random number between 0 to 9
print(random.randint(1,6)) # It generate a random value from 1 to 6
print(random.randint(1000,9999))   # For OTP purpose
print(random.choice(['A',1,2,3,4,5,6,7,8,9,'J','Q','K']))   # This function returns the random value from given sequence
print(random.choices(['A',1,2,3,4,5,6,7,8,9,'J','Q','K'],k=3))   # This function returns the random values from given sequence based on the k value
arr = ['A',1,2,3,4,5,6,7,8,9,'J','Q','K']
random.shuffle(arr)        # It will shuffle the values
print(arr)
```

```python
rolls = 0
count = 0
a = int(input())
while(count<a):
    input("Press Enter to Throw the Die")
    ran = random.randint(1,6)
    rolls += 1
    if count+ran <= a:
        count += ran
        print(f"Throw : {rolls}")
    print(f"Throw Value : {ran}")
    print(f"Your total After throw : {count}")
print(f"Total Rolls : {rolls}")
    
```
