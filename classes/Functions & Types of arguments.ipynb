# Functions:

### Function :

            A function is reusable block of code in python. function definition starts with 'def' keyword.

            - Syntax :
                    
                    #function defination block
                    def fun_name(parameters:(Optional)):
                            # block code 
                            return exp/value (Optional)

                    #Function calling block
                    fun_name(arguments(Optional))

### Types of function :

            1. User defined functions.
            2. Builtin functions.

### 1.User defined function :

            The function definition is developed by the user or programmer , those functions are called as "user defined functions".

            - A function is used to perform a specific task.
            - The function definition block executes when function call is occured.

    - Parameters :

            The function definition block takes variables as input those variables are defined inside the paranthasis, those variables called as parameters.
            -Paramaters are optional.

    - Arguments :

            The function calling block takes variables as input those variables are defined inside the paranthasis, those variables called as arguments.
            -Arguments are optional.

        Based on parameters and return value functions are classified into 4 types.

                        1. Without parameters and without return.
                        2. without parameters and with return.
                        3. with parametrs and without return.
                        4. with parameters and with return.

```python
# Types of functions based on parameters and return value
# 4. with parameters and with return.
# function defination block
def addition(x:int,y:int)->int:
    c = x+y
    return c
# Function calling block
a,b = 10,20
res = addition(a,b)
print(res)
# addition(10,20)
```

```python
# 3. with parameters and without return.
# by default function returns None value, when function not returning any value
# function defination block
def addition(x:int,y:int)->int:
    c = x+y
    print(f"This is from function defination block:{c}")
# Function calling block
a,b = 10,20
res = addition(a,b)
print(res)
print(f"this is from function calling block : {res}")
# addition(10,20)
```

```python
# 2. without parameters and with return.
# function defination block
def addition()->int:
    a = 10
    b = 20
    return a+b
    print(f"This is from function defination block:{c}")
# Function calling block
res = addition()
print(res)
print(f"this is from function calling block : {res}")
# addition(10,20)
```

```python
# 1. without parameters and without return.
# function defination block
def addition()->None:
    a = 20
    b = 20
    c = a+b
    print(f"This is from function defination block:{c}")
    print("This is from function defination block")
# Function calling block
res = addition()
print(res)
print(f"this is from function calling block : {res}")
# addition(10,20)
```

```python
# even or odd using functions
def even_or_odd(x:int):
    if x%2 == 0:
        return "Even"
    else:
        return "odd"
a = int(input())
even_or_odd(a)
```

```python
# even or odd using functions
def prime_number(x:int):
    for i in range(2,x//2+1):
        if x%i == 0:
            return "Not prime"
    return "Prime"
a = int(input())
prime_number(a)
```

# Types of arguments :

        There are 5 types of arguments. they are :

                    1. Positional arguments
                    2. keyword arguments
                    3. Default arguments
                    4. Variable length of positional arguments
                    5. Variable length of keyword arguments

```python
#1.Positional Arguments
def sub(x:int,y:int)->int:
    z = x-y
    return z
a = 10
b = 5
print(sub(a,b))
print(sub(b,a))
```

```python
# 2. keyword arguments : arguments are explicitly passed with parameter names
def sub(x:int,y:int)->int:
    z = x-y
    return z
a = 10
b = 5
print(sub(x = a,y = b))
print(sub(y = b,x = a))
```

```python
# 2. keyword arguments : arguments are explicitly passed with parameter names
def add(x:int,y:int)->int:
    z = x+y
    return z
a = 10
b = 5
print(add(x = a,y = b))
print(add(y = b,x = a))
# print(add(10,x=5))                  # error - add() got multiple values for argument 'x'
# print(add(x=5,10))                  # error - positional argument follows keyword argument
```

```python
# sum of 3 numbers
def add(a:int,b:int,c:int)->int:
    sum = a+b+c
    return sum
x = 10
y = 5
z = 5
print(add(a = x,b = y,c = z))
print(add(x,b = y,c = z))
```

```python
# sum of 4 numbers
# 3. Default arguments
def add(a:int,b:int,c:int = 0,d:int = 0)->int:
    sum = a+b+c+d
    return sum
x,y,z,w = map(int,input().split())
print(add(a = x,b = y,c = z,d = w))
print(add(x,y))
print(add(x,y,z))
print(add(x,y,z,w))
```

```python
# Multiplication of 4 numbers
# 3. Default arguments
def add(a:int,b:int,c:int = 1,d:int = 1)->int:
    mul = a*b*c*d
    return mul
x,y,z,w = map(int,input().split())
# print(add(a = x,b = y,c = z,d = w))
print(f"Multiplicatio of 2 numbers : {add(x,y)}")
print(f"Multiplicatio of 3 numbers : {add(x,y,z)}")
print(f"Multiplicatio of 4 numbers : {add(x,y,z,w)}")
```

```python
#  4. Variable length of positional arguments
def add(*args):
    s = 0
    for i in args:
        s += i
    return s
x,y,z,w = map(int,input().split())
# print(add(a = x,b = y,c = z,d = w))
print(f"Multiplicatio of 1 numbers : {add(x)}")
print(f"Multiplicatio of 2 numbers : {add(x,y)}")
print(f"Multiplicatio of 3 numbers : {add(x,y,z)}")
print(f"Multiplicatio of 4 numbers : {add(x,y,z,w)}")
```

```python
#  4. Variable length of keyword arguments
def add(*args):
    s = 0
    for i in args:
        s += i
    return s
lst = list(map(int,input().split()))
# print(add(a = x,b = y,c = z,d = w))
print(f"Multiplicatio of 1 numbers : {add(*lst)}")
print(*lst)
print(lst)
```

```python
# in python every type function calling blocks shares the arguments refference to parameters
# if the arguments are immutable data types, then behave likes pass by value and if arguments are mutable datatypes(ie, list,set,dict) then behave like pass by referrence.
# pass by value
def fun(x):
    print(id(x))
    x += 2
    print("After adding 2")
    print(id(x))
    print(x)
a = 10
fun(a)
print(id(a))
print(a)
```

```python
# pass by reference
def add_ele(x,y):
    x.append(y)

arr = [1,2,3,4]
print("Before arr value",arr)
add_ele(arr,100)
print("After arr value",arr)
```
