# List Comprehension :

                This is the one of the way to create a list with an expression

                Syntax :

                    [expression for iterator in iterable if condition]

```python
# finding 1 to n even numbers
n = int(input())
[i for i in range(1,n+1) if i%2==0]
```

```python
# finding n even numbers
n = int(input())
[i*2 for i in range(1,n+1)]
```

```python
# finding the first n quibic numbers
n = int(input())
[i**3 for i in range(1,n+1)]
```

```python
# finding the odd numbers between the range
n1,n2 = map(int,input().split())
[x for x in range(n1,n2+1) if x%2 == 1]
```

```python
import math
arr = list(map(int,input().split()))
[x**2 if x%2==0 else round(math.sqrt(x),2) for x in arr]
```

```python
# find the 2 die combinaton values
[(x,y) for x in range(1,7) for y in range(1,7)]
```

```python
# return the combinations is divisible by 3
[(x,y) for x in range(1,7) for y in range(1,7) if (x+y)%3==0]
```

```python
# return the combinations is divisible by 3
[(x,y) for x in range(1,7) for y in range(1,7) if x == 4 or y == 4]
```
