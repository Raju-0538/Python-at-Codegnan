### File :

        The file is used to store data in electronic format in system.

### File Operations:

        1. Opening a file -------------->open()
        2. Read a file------------------>read()
        3. Writing into to file--------->write()
        4. closing file----------------->close()

## 1. Opening a File:

        To open a file by using open() function in that we need to pass file name and file opening mode.

        Syntax:
                file_obj = open(file_name,mode)
                # read/ write operations
                # close file

        file opening modes:
                1. r - Opening a file in Read mode. if file not exists it raises 'file not found' error.
                2. w - Opening a file in Write mode. if file not exists it will create a new file. if file is exists it will arase all data in the                            file and writes the new data. 
                3. a - Opening a file in append mode. if file is not exists it will create new file. if file exists it will append the new data to                            previous data.
                4. x - It is used to create a new file.
                5. b - This mode is used to perform operations on binary files(ex:images) for writing we can use wb mode and for reading we can use rb                        mode.
                6. r+ - Opening a file in Read + Write mode.
                7. w+ - Opening a file in Write + Read Mode.
                8. a+ - Opening a file in Append + Read mode.

## 2. Reading a content from a file:

        To perform the read operations on file for that we need to open file in 'r' mode and then we can perform following operations.
                
                1. read() : It returns entire file content as single string.
                2. readline() : This method returns first line as strings.
                3. readlines() : It returns all the lines as list.

## 3. Writing a content into to a file:

        To perform the writing operations on file for that we need to open file in 'w' mode and then we can perform following operations.
                
                1. write(string) : It adds string into to a file.
                2. writelines(list of lines) : It adds all list lines to a file.

## 4. Closing a file :

        After performing the read or write operations we need to close the file by using close() function. Otherwise there is a chance of data loss.
